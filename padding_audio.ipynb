```python
# Padding audio data

# Importing required libraries
import librosa
import numpy as np
import os

# Loading and flattening siri audio files
for file in os.listdir("siri_audio_assets"):
    if not ".DS_Store" in file:
        audio_file = f"siri_audio_assets/{file}"
        y, sr = librosa.load(audio_file) # y is amplitude, sr is frequency
        y = np.squeeze(y) # flattening the audio files
    
# Loading and flattening not_siri audio files
for file in os.listdir("not_siri_audio_assets"):
    if not ".DS_Store" in file:
        audio_file = f"not_siri_audio_assets/{file}"
        y, sr = librosa.load(audio_file) # y is amplitude, sr is frequency
        y = np.squeeze(y)

print(f"Sample rate: {sr}")

print("=================================== Padding siri audio data =======================================")
print("\n")

temp = []

# Padding the siri audio data
for file in os.listdir("siri_audio_assets"):
    if not ".DS_Store" in file:
        if not ".ipynb_checkpoints" in file:
            arr,_ = librosa.load(f"siri_audio_assets/{file}")
            # print(f"---------------- Unpadded array: {arr.shape}")
            temp.append(arr.shape[0])     
            if arr.shape[0] < 76032:
                arr_padded = np.pad(arr, (0,abs(arr.shape[0]-76032)), "constant")
                # Saving the siri padded audio data
                np.save(file[:len(file)-4],arr_padded,allow_pickle=False,fix_imports=False)
                print("Saved padded Siri Data")
                
            # Saving Siri data that does not need to be padded
            if arr.shape[0] == 76032:
                np.save(file[:len(file)-4],arr,allow_pickle=False,fix_imports=False)
                print("Saved Siri Data")


print("\n")
print("=================================== Now padding not_siri audio data =======================================")
print("\n")

# Padding the not_siri audio data
for file in os.listdir("not_siri_audio_assets"):
    if not ".DS_Store" in file:
        if not ".ipynb_checkpoints" in file:
            arr,_ = librosa.load(f"not_siri_audio_assets/{file}")
            temp.append(arr.shape[0])     
            # print(f"---------------- Unpadded array: {arr.shape}")
            if arr.shape[0] < 76032:
                arr_padded = np.pad(arr, (0,abs(arr.shape[0]-76032)), "constant")
                # Saving the not_siri padded audio data
                np.save(file[:len(file)-4],arr_padded,allow_pickle=False,fix_imports=False)
                print("Saved padded not_siri Data")
                      
            # Saving not_siri data that does not need to be padded
            if arr.shape[0] == 76032:
                np.save(file[:len(file)-4],arr,allow_pickle=False,fix_imports=False)
                print("Saved not_siri Data")
            
print(f"Max: {max(temp)}")
```
